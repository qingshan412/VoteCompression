```python
import numpy as np
np.random.seed(123)
init_si = 20
init_ci = 10
x = np.random.rand(init_si,init_ci)
#print(x)
w = np.random.rand(init_ci,1)
#print(w)
y = np.dot(x,w)
#print(y)
ci = init_ci
si = init_si
Dr = ci-si
print(Dr)#
Dr_list = []
Dis_list = []
for i in range(init_si - 1): 
    si = i + 2 
    xi = x[:si, :]
    yi = y[:si, :]
    #[Dr, Dis] = FindMinD(xi, yi)
    Dr = ci - si
    Bin = 2**ci
    Dis = 10000
    for j in range(Bin):
        wi_b = list(bin(j))[3:]
        wi_bit = [2*int(i)-1 for item in wi_b]
        wi = np.array(wi_bit).reshape((-1, 1))
        if len(wi_bit)<ci:
            wi = np.vstack((wi, -np.ones((ci-len(wi_bit), 1))))
        y_h = np.dot(xi, wi)
        Dis_t = np.sum(np.square(yi - y_h))
        if Dis_t < Dis:
            Dis = Dis_t
    print(str(Dr) + ':' +str(Dis/si))
    Dr_list.append(Dr)
    Dis_list.append(Dis/si)
```

```python
%matplotlib notebook
#Dis_list[4] = Dis_list[5]
Dr_use = Dr_list[7:-6]
Dis_use = Dis_list[7:-6]
print(Dr_use)
print(Dis_use)
import matplotlib.pyplot as plt
plt.plot(Dr_use, Dis_use, '-o')
```
